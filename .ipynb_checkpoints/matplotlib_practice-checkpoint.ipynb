```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
ax = plt.axes()
ax.plot(x_axis, y_axis, label='Boston', marker="*", color="green", linewidth=2)
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
ax.set_xlim(0, 11)
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add a grid 
ax.grid()
# Add the legend.
ax.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create scatter plot
# plt.plot(x_axis, y_axis, 'o')
plt.scatter(y_axis, x_axis, c="red", label="Chicago")
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.xlim(0, 50)
plt.gca().invert_yaxis()
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])

# Increase size of the markers 
#y_axis_larger = []
#for data in y_axis:
  #y_axis_larger.append(data*3)
```

```python
# Create a scatter plot using the object-oriented interface method 
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s= [i * 5 for i in y_axis], label="Boston", c="skyblue", edgecolors="black", linewidth=2)
# Create labels for the x and y axes.
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.invert_yaxis()
ax.set_xlim(0, 50)
# Create a title.
ax.set_title("PyBer Fare by Month")
ax.legend()
```

```python
# Create pie chart
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Create pie chart using the object-oriented interface method
fig, ax = plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis,labels=x_axis, explode=explode_values, colors=colors, autopct='%.1f%%', shadow=True, startangle=95, counterclock=False)

plt.show()
```
